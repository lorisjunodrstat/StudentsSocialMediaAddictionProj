```r
dir()
SSMA_link <- 'SDSMA_Students Social Media Addiction.csv'
```

```r
SSMA <- read.csv2(SSMA_link, sep=",", header = T)
```

```r
# Clean the data (character)
SSMA$Gender <- as.factor(SSMA$Gender)
SSMA$Country <- as.factor(SSMA$Country)
Academic_order <- c("High School", "Undergraduate", "Graduate")
SSMA$Academic_Level <- factor(SSMA$Academic_Level, levels=Academic_order)
SSMA$Avg_Daily_Usage_Hours <- as.numeric(SSMA$Avg_Daily_Usage_Hours)
SSMA$Most_Used_Platform <- as.factor(SSMA$Most_Used_Platform)
SSMA$Affects_Academic_Performance <- as.factor(SSMA$Affects_Academic_Performance)
SSMA$Sleep_Hours_Per_Night <- as.numeric(SSMA$Sleep_Hours_Per_Night)
SSMA$Relationship_Status <- as.factor(SSMA$Relationship_Status)

```

```r
library(ggplot2)
```

```r
ggplot(SSMA, aes(x = Avg_Daily_Usage_Hours, y = Mental_Health_Score)) + geom_point() + facet_grid(~Relationship_Status)
```

```r

ggplot(SSMA, aes(x = Avg_Daily_Usage_Hours, y = Mental_Health_Score)) + geom_point() + facet_grid(~Most_Used_Platform)
ggplot(SSMA, aes(x = Avg_Daily_Usage_Hours, y = Mental_Health_Score)) + geom_point() + facet_grid(~Academic_Level)
ggplot(SSMA, aes(x = Avg_Daily_Usage_Hours, y = Mental_Health_Score)) + geom_point() + facet_grid(~Affects_Academic_Performance)
ggplot(SSMA, aes(x = Avg_Daily_Usage_Hours, fill = Affects_Academic_Performance)) + geom_density(alpha =0.5)
ggplot(SSMA, aes(x = Avg_Daily_Usage_Hours, fill = Most_Used_Platform)) + geom_density(alpha =0.5)
ggplot(SSMA, aes(x = Avg_Daily_Usage_Hours, fill = Academic_Level)) + geom_density(alpha =0.5)
```
